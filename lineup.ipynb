```python
import matplotlib.pyplot as plt
from mplsoccer import Pitch

# Create horizontal pitch
pitch = Pitch(pitch_type='opta')
fig, ax = pitch.draw()

# Define 4-3-3 formation positions (x, y coordinates) and player names
positions = [
    (25, 15, 'Right back'), (20, 35, 'Right centreback'), (20, 65, 'Left centreback'), (25, 85, 'Left back'),   # Back four
    (40, 35, 'Right central midfielder'), (30, 50, 'Defensive midfielder'), (40, 65, 'Left central midfielder'), # Midfield three
    (60, 20, 'Right wing'), (65, 50, 'Central striker'), (60, 80, 'Left wing')                                  # Front three
]

# Plot players with text below the dots
for x, y, name in positions:
    pitch.scatter(x, y, ax=ax, s=100, color='blue')
    ax.text(x, y + 5, name, ha='center', va='top', color='black', fontsize=6)

# Goalkeeper with text below the dot
pitch.scatter(2, 50, ax=ax, s=100, color='blue')
ax.text(2, 50 + 5, 'Goalkeeper', ha='center', va='top', color='black', fontsize=6)

plt.show()
```
